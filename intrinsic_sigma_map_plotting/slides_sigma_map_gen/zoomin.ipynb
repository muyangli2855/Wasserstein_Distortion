```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
filename='taipei'
img = np.array(Image.open(filename+'.jpg'))
img_recon = np.array(Image.open(filename+'_reproduction.png'))
img_smap = np.load(filename+'_sigma_map.npy')
```

```python
plt.figure(figsize=(8,8))
plt.imshow(img)
plt.axis('off')
plt.plot(268, 340, marker='o', color="blue",markersize=3) 
plt.plot(267, 350, marker='o', color="yellow",markersize=3) 
plt.plot(268, 370, marker='o', color="green",markersize=3) 
plt.plot(270, 380, marker='o', color="red",markersize=3) 
```

```python
print(img_smap[340,268],img_smap[350,267],img_smap[370,268],img_smap[380,270])
```
