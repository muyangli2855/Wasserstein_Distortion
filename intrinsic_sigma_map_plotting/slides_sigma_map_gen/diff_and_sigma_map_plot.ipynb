```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

```python
filename='writing'
img = np.array(Image.open(filename+'.jpg'))
# img_smap = np.load(filename+'_sigma_map_old.npy')
img_recon = np.array(Image.open(filename+'_reproduction.png'))
```

```python
diff = np.mean(img-img_recon,axis=2)
plt.figure(figsize=(8,8))
plt.imshow(diff,cmap='gray')
plt.axis('off')
plt.savefig(filename+'_diff.png',bbox_inches='tight')
```
