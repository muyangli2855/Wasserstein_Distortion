```python
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
```

```python
zebra = np.array(Image.open('exp4/zebra.jpg'))
zebra_sigma = np.array(Image.open('sigmaMap/zebra_sigma.png'))
zebra_recon = np.array(Image.open('Reproduction/zebra_reproduction.png'))

zebra_sali  = np.sqrt(np.array(Image.open('exp4/zebra_sali.png').convert('L'))/255.,dtype=np.float32)
zebra_sigma_from_Sali = np.array(Image.open('exp3/zebra_sigma.png'))
zebra_recon_from_Sali = np.load('exp3/zebra.npy')

usopen = np.array(Image.open('exp4/usopen.jpg'))
usopen_sigma = np.array(Image.open('sigmaMap/usopen_sigma.png'))
usopen_recon = np.array(Image.open('Reproduction/usopen_reproduction.png'))

usopen_sali  = np.sqrt(np.array(Image.open('exp4/usopen_sali.png').convert('L'))/255.,dtype=np.float32)
usopen_recon_from_Sali = np.load('exp3/usopen.npy')
usopen_sigma_from_Sali = np.array(Image.open('sigmaMap/usopen_sigma_from_Sali.png'))
```

```python
images = [zebra,usopen]
images_recon = [zebra_recon,usopen_recon]
images_sigma = [zebra_sigma,usopen_sigma]
image_sali = [zebra_sali, usopen_sali]
images_recon_from_Sali = [zebra_recon_from_Sali,usopen_recon_from_Sali]
images_sigma_from_Sali = [zebra_sigma_from_Sali,usopen_sigma_from_Sali]
for i in range(2):
    fig,ax = plt.subplots(2,3,figsize=(18,12))
    plt.subplots_adjust(wspace=0, hspace=.02) 
    
    ax[0][0].imshow(images[i])
    ax[0][0].axis('off')
    ax[0][0].set_title('Source',fontsize=18)

    ax[0][1].imshow(images_sigma[i],cmap='gray',extent=[0,480,480,0],aspect='auto')
    ax[0][1].imshow(images[i],alpha=0.2)
    ax[0][1].axis('off')
    ax[0][1].set_title(r'$\sigma$ Map',fontsize=18)

    ax[0][2].imshow(images_recon[i])
    ax[0][2].axis('off')
    ax[0][2].set_title('Reconstruction',fontsize=18)

    ax[1][0].imshow(image_sali[i], cmap='gray')
    ax[1][0].axis('off')


    ax[1][1].imshow(images_sigma_from_Sali[i],cmap='gray',extent=[0,480,480,0],aspect='auto')
    ax[1][1].imshow(images[i],alpha=0.2)
    ax[1][1].axis('off')

    ax[1][2].imshow(images_recon_from_Sali[i])
    ax[1][2].axis('off')

    
    fig.savefig('metamer_{}.png'.format(i+1),bbox_inches = 'tight',pad_inches = 0)
```

```python
images = [sheepdog, ski, usopen2, taipei, hydrant, horses, fu, mona_lisa, rubber_soul, words,lego_faces]
images_sigma = [sheepdog_sigma, ski_sigma, usopen2_sigma, taipei_sigma, hydrant_sigma, horses_sigma, fu_sigma, mona_lisa_sigma, rubber_soul_sigma, words_sigma,lego_faces_sigma]
images_recon = [sheepdog_recon, ski_recon, usopen2_recon, taipei_recon, hydrant_recon, horses_recon, fu_recon, mona_lisa_recon, rubber_soul_recon, words_recon,lego_faces_recon]

fig, axes = plt.subplots(4, 3, figsize=(18, 30))
plt.subplots_adjust(wspace=0, hspace=.02) 


for i in range(4):
    # Source image
    i = i 
    axes[i, 0].imshow(images[i])
    axes[i, 0].axis('off')
    if i == 0:
        axes[i, 0].set_title('Source', fontsize=18)

    # Sigma map with overlay
    axes[i, 1].imshow(images_sigma[i], cmap='gray', extent=[0, 480, 480, 0], aspect='auto')
    axes[i, 1].imshow(images[i], alpha=0.2)
    axes[i, 1].axis('off')
    if i == 0:
        axes[i, 1].set_title(r'$\sigma$ Map', fontsize=18)

    # Reconstruction
    axes[i, 2].imshow(images_recon[i])
    axes[i, 2].axis('off')
    if i == 0:
        axes[i, 2].set_title('Reconstruction', fontsize=18)

plt.savefig('image1.png', bbox_inches='tight')
plt.close(fig)
```
